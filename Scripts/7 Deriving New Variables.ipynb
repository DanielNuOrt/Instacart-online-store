### Create new variable using 'def' with conditional function 'if, elif, else'.

```python
# using 'def' with conditional function 'if, elif, else'.
def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High-range'
    else: 
        return 'Not enough data'
```

```python
#Creating new variable 'price_range'
df['price_range'] = df.apply(price_label, axis = 1)
```

```python
# .value_counts() count of entries with the same values in a column
df['price_range'].value_counts(dropna = False)
```

### If-Statements with For-Loops

```python
# Categorize the day of the week depending of the number of orders using For_loop
import time
start_time = time.time()

result = []

for value in df_ords_prods["orders_day_of_the_week"]:
  if value == 'Saturday':
    result.append("Busiest day")
  elif value == 'Wednesday':
    result.append("Least busy")
  else:
    result.append("Regularly busy")
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
result
```

```python
df_ords_prods['busiest_day'] = result
```

```python
# Check value frecuency in "busiest_day" 
df_ords_prods['busiest_day'].value_counts(dropna = False)
```

# 02. Create 'busiest_days' column

```python
# Create a new colums. If-Statements with For_Loops. “Busiest days” (plural).
# Using '.time' function for fun.
start_time = time.time()

result2 = []

for value in df_ords_prods["orders_day_of_the_week"]:
  if value == 'Saturday' or value == 'Sunday':
    result2.append("Busiest days")
  elif value == 'Wednesday' or value =='Tuesday':
    result2.append("Slowest days")
  else:
    result2.append("Regularly busy")
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
# Adding result2 to a new column "busiest_ days"
df_ords_prods['busiest_days'] = result2
```

```python
df_ords_prods['busiest_days'].value_counts(dropna = False)
```

```python
# Identify the busiest hours of the day.
df_ords_prods['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.”
# Create a new column containing these labels called “busiest_period_of_day.”
result_per_hours = []

for value in df_ords_prods["order_hour_of_day"]:
  if value >= 9 and value <= 16:
    result_per_hours.append("Most orders")
  elif value == 23 or (value <= 6 and value >= 0):
    result_per_hours.append("Fewest orders")
  else:
    result_per_hours.append("Average orders")
```

```python
# Creating a new column "busiest_period_of_day"
df_ords_prods['busiest_period_of_day'] = result_per_hours
```
